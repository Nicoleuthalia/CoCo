### Task 1

Graph the Gaussian function with at least three different values of  ùúé .

```python
import numpy as np
import matplotlib.pyplot as plt

#Definiendo el arreglo de mis valores de ùúé
sigma = np.array([1., 2., 3., 4., 5.])

#Definiendo la variable independiente
x = np.linspace(-100,100,1001)

#Definiendo la variable dependiente g(x) para diferentes valores de ùúé
g1 = (1/(sigma[0]*np.sqrt(2*np.pi)))*np.exp((-0.5)*(x/sigma[0])**2)
g2 = (1/(sigma[1]*np.sqrt(2*np.pi)))*np.exp((-0.5)*(x/sigma[1])**2)
g3 = (1/(sigma[2]*np.sqrt(2*np.pi)))*np.exp((-0.5)*(x/sigma[2])**2)
g4 = (1/(sigma[3]*np.sqrt(2*np.pi)))*np.exp((-0.5)*(x/sigma[3])**2)
g5 = (1/(sigma[4]*np.sqrt(2*np.pi)))*np.exp((-0.5)*(x/sigma[4])**2)

#Graficar mis diferentes funciones
plt.plot(x, g1, linestyle="-", label="$\sigma$ = 1")
plt.plot(x, g2, linestyle="-.", label="$\sigma$ = 2")
plt.plot(x, g3, linestyle="--", label="$\sigma$ = 3")
plt.plot(x, g4, linestyle="-", label="$\sigma$ = 4")
plt.plot(x, g5, linestyle=":", label="$\sigma$ = 5")

plt.legend()
plt.title('Funci√≥n Gaussiana Centrada en Cero')
plt.xlabel("x")
plt.ylabel("g(x)")
plt.xlim([-20, 20])
plt.grid()
plt.savefig("funciongaus.png", dpi=300)
plt.show()
```
